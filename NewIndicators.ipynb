#### 6. Tasa de rendimiento (créditos aprobados / matriculados). Se puede especificar por tiempo
1. semestre consta de 215 creditos en total, en este caso
2. (Matriculados)total de creditos que se a escrito una vez terminado el semestre,
3. sacar indicador por semestre, aprobados / matriculados
4. jemplo llevo 30 creditos y aprobo 15

```python
Rendimiento = NewData[(NewData["Matricula"] == 'E14080632') & (NewData["Programa de estudios"] == "ISIC-2010-224-INGENIERÍA EN SISTEMAS COMPUTACIONALES") &
                      (NewData["Ciclo escolar de la materia"]%2 != 0)]
Rendimiento
```

```python
#Sacar primedio de creditos por semestre de manera manual
#no incluir verano
Rendimiento[(Rendimiento["Semestre"]==1)][["Creditos"]].sum()
```

```python
#Sacar el promedio de cada semestre
creditosSuma = Rendimiento.groupby(['Semestre'])[['Creditos']].sum()
creditosSuma["Aprovados/Matriculados"] = creditosSuma/28
creditosSuma
```

```python
creditosSuma['Aprovados/Matriculados'].sum()
```
